```python
import pandas as pd
```

```python
pangkat = pd.read_csv('pangkat.csv')
```

```python
data = pangkat.copy()
```

```python
target = 'pangkat'
```

```python
encode = ['pend_terakhir']
```

```python
data
```

```python
for col in encode:
  dummy = pd.get_dummies(data[col], prefix=col)
  data = pd.concat([data,dummy], axis=1)
  del data[col]
```

```python
data
```

```python
target_mapper = {'I/d':0, 'II/a':1, 'II/b':2, 'II/c':3, 'II/d':4, 'III/a':5, 'III/b':6,
                 'III/c':7, 'III/d':8, 'IV/a':9, 'IV/b':10, 'IV/c':11, 'IV/d':12}
```

```python
def target_encode(val):
  return target_mapper[val]
```

```python
data['pangkat'] = data['pangkat'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('pangkat', axis=1)
Y = data['pangkat']
```

```python
X
```

```python
Y
```

```python
# MODEL NAIVE BAYES CLASSIFIER (NBC)
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,Y)
```

```python
# MENYIMPAN MODEL NAIVE BAYES CLASSIFIER (NBC)
import pickle
pickle.dump(model, open('modelNBC_pangkat.pkl', 'wb')) #wb untuk menyimpan
```
